###1. Utilice la Función de Intensidad vista en clase.


```r
fdi <- function(t) {
  
  if( t <= 3){
    v <- 5
  }
  
  if( t > 3 && t <= 6){
    v <- ((10/3)*t)-5
  }
  
  if( t > 6 && t <= 8){
    v <- (-7*t)+57
  }
  
  if(t > 8){
    v <- 1
  }
  
  return(v)
}
```


###2. Suponga que la llegada de clientes sigue un Proceso Poisson no Homogéneoque es generado a partir de un Proceso Poisson Homogéneo con λ= 15.


```r
lambda <- 15

Ts <- function(s,lambda){
  
  t <- s
  terminar <- 0
  while(terminar == 0){
    u <- runif(1)
    t <- t -(1/lambda)*log(u)
    u <- runif(1)
    if( u <= (fdi(t)/lambda) ){
      
      ts <- t
      terminar <- 1
    }
  }
  
  return(ts)
}
```
